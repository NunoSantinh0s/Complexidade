#### Exemplo 1 de boxplot

```python
# criação de um boxplot 

from matplotlib import pyplot as plt

tempos = [15,13,12,15,16,31,14,15,17,14,16,14,18,12,12,15,20,7]

plt.boxplot(tempos)
plt.xlabel('valor do input n da função f(n)')
plt.ylabel('tempo de execução da função f(n) [s]')
plt.title("Estudo da variabilidade do tempo de execução\n da função f para o valor n")

plt.show()
```

#### Exemplo 2 de boxplot

```python
# dicas sobre boxplot 

from matplotlib import pyplot as plt

dic_tempos = {'20': [34.54, 34.345, 34.761], '40': [34.541, 34.748, 34.482]}

fig, ax = plt.subplots()
ax.boxplot(dic_tempos.values())
ax.set_xticklabels(dic_tempos.keys())
ax.set_xlabel('valor do input n da função f(n)')
ax.set_ylabel('tempo de execução da função f(n) [s]')
ax.set_title("Estudo da variação do tempo de execução \nda função f(n), para vários valores de n")

ax.legend(dic_tempos.keys())


plt.show()
```

### Ex. 2.3: minimo simples 

* ver slides
* crie como input uma lista com valores aleatorios, usando a função rand.

```python
# minimo simples
import random

def minimosimples():
    n = int(input('Lista: '))
    lista = []

    for i in range(n):
        numero_aleatorio = random.randint(1,100)
        lista.append(numero_aleatorio)

    minimo = lista[0]
    for num in lista:
        if num < minimo:
            minimo = num
    
    return minimo

```

### Ex. 2.4  minimo quadratico

* ver slides
* crie como input uma lista com valores aleatorios, usando a função rand.


```python
# minimo quadrático

def minimoQuadratico():
    n = int(input('QUantos numeros na lista: '))
    lista = []

    for _ in  range(n):
        numero_aleatorio = random.randint(1, 100)
        lista.append(numero_aleatorio)
    
    minimo = lista[0]

    for e in lista[0]:
        for f in lista:
            if e < f and e < minimo:
                minimo = e
    
    return minimo


```
